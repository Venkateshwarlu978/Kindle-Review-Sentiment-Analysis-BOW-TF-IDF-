```python
##clean the data
#1.removing special characters(stopwards etc..)
#special characters
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-z 0-9]+','',x))
#stopwords
df['reviewText']=df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
#remove url
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(r'http\S+|www\S+', '', x))
#remove html tags
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(r'<.*?>', '', x))
#remove any additional spaces
df['reviewText'] = df['reviewText'].str.replace(r'\s+', ' ', regex=True).str.strip()
```
